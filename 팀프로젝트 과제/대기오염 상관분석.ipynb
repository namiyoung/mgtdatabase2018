```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# 중위단위매매가격 price 변수 설정
price = pd. read_csv("price.csv", encoding = "euc-kr")
price.head()
```

```python
# 중위단위매매가격 2017년 12월 기준 오름차순 정렬
price.sort_values(by="2017년 12월", ascending = False)
```

```python
# 서울시 구별 대기오염(2017년 데이터) air_pollution 변수 설정
air_pollution = pd. read_csv("air pollution.csv", encoding = "euc-kr")
air_pollution.head()
```

```python
# 서울시, 구별 대기오염 초미세먼지 기준 오름차순 정렬
air_pollution.sort_values(by="초미세먼지", ascending = False)
```

```python
# data_result = 매매가격 & 초미세먼지 대기오염 병합
data_result = pd.merge(price,air_pollution,on="구별")
data_result
```

```python
# data_result = 구별로 중위단위매매가격 2017년 12월 &  대기오염 초미세먼지
data_result = data_result.loc[:,["구별","2017년 12월","초미세먼지"]]
```

```python
data_result.head()
```

```python
data_result.describe()
```

```python
np.corrcoef(data_result["2017년 12월"],data_result["초미세먼지"])
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="2017년 12월", y="초미세먼지", data=data_result, size=7) # regression plot
plt.show()
```
